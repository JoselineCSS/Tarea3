```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# La misma función para definir una distribución gaussiana utilizada en Pregunta3.ipynb
def gaussian(domain, sigma, mu):
    return (1/(np.sqrt(2*np.pi*sigma**2)))*np.exp( (-(domain-mu)**2)/(2*sigma**2) )

# Se definen "preimágenes" para los valores de X y Y con 1000 puntos, para que se asemejen a una PDF
X = np.linspace(5, 15, 1000)
Y = np.linspace(5, 25, 1000)
# Se obtienen las distribuciones normales asociadas a X y Y
xModel = gaussian(X, 3.3, 9.905) 
yModel = gaussian(Y, 6.06, 15.08)

# Se grafican los modelos teóricos continuos de las
# probabilidades marginales de X y Y. Se nota que sí producen distribuciones normales
plt.plot(X, xModel, alpha=0.75)
plt.grid(axis="y", alpha=0.7)
plt.xlabel("Valor de X")
plt.ylabel("Frecuencia relativa de X")
plt.title("Distribución de X")
# plt.savefig("img/xTheoCurve.png")
plt.show()

plt.plot(Y, yModel, alpha=0.75)
plt.grid(axis="y", alpha=0.7)
plt.xlabel("Valor de Y")
plt.ylabel("Frecuencia relativa de Y")
plt.title("Distribución de Y")
# plt.savefig("img/yTheoCurve.png")
plt.show()

# Se utiliza el toolkit de graficación en 3D de matplotlib para graficar la probabilidad conjunta de X y Y.
# Se utiliza un wireframe, que necesita que X y Y formen un "mesh grid, una malla"
X, Y = np.meshgrid(X,Y)
# El valor teórico de f_{X,Y}(x,y) definido en la asignación 2
Z = (1/(2*np.pi*19.9))*np.e**( -(((X-9.905)**2)/21.78) - (((Y - 15.08)**2)/72.72) )

# El resto es para crear la gráfica tridimensional. Definición de los ejes, sus etiquetas, etc.
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z, color="b")
ax.set_xlabel("Valor de X")
ax.set_ylabel("Valor de Y")
ax.set_zlabel("Probabilidad conjunta de X y Y")
ax.set_title("Distribución conjunta de X y Y")
ax.view_init(30, 35)
# plt.savefig("img/zTheoCurve.png")
plt.show()
```
